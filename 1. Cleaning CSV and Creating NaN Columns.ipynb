```python
import os
import pandas as pd
from transformers import pipeline
import numpy as np
```

```python
df = pd.read_csv('../data/sentiment_analysis_data.csv', encoding='utf-8')
print("Length of df with NaN texts:", len(df))
df = df.drop(columns=['sentiment'])
df = df.loc[~df['text'].isnull()]
print("Updated length of df without NaN texts:", len(df))
df.head()
```

```python
df['subjectivity'] = [np.nan]*len(df)
df['polarity'] = [np.nan]*len(df)
df['emotion'] = [np.nan]*len(df)
df.head()
```

```python
df.to_csv('../data/step1.csv', index=False)
```
